```python
import re
def startsWithDate(s):
    pattern='\[([0-2][0-9]|(3)[0-1])\/(((0)[1-9]|(1)[0-2]))\/(\d{2}|\d{4}), ((\d|(1)[0-2])\:([0-5][0-9])\:([0-5][0-9]) ([A|P][M])\])'
    result = re.match(pattern, s)
    if result:
        return True
    return False
```

```python
def getDTAM(line):
    date=line.split(']')[0].split('[')[1].split(',')[0]
    time=line.split(']')[0].split('[')[1].split(',')[1].strip()
    if len(time)==10:
        time='0'+time
   
    if len(line.split(']')[1].split(':'))>1:
        author=line.split(']')[1].split(':')[0].strip()
        if len(line.split(']')[1].split(':'))>2:
            s=':'
            msg=s.join(line.split(']')[1].split(':')[1:])
        else:
            msg=line.split(']')[1].split(':')[1].strip()
    else:
        author=None
        msg=line.split(']')[1].split(':')[0].strip()
    return date,convertTo24(time).strip(),author,msg
```

```python
def convertTo24(str1):  
    if str1[-2:] == "AM" and str1[:2] == "12": 
        return "00" + str1[2:-6] 
    
    elif str1[-2:] == "AM": 
        return str1[:-6] 
    
    elif str1[-2:] == "PM" and str1[:2] == "12": 
        return str1[:-6] 
    
    else:
        return str(int(str1[:2]) + 12) + str1[2:5]
```

```python
count1=0
count2=0
count3=0
count4=0
count5=0
count6=0
count7=0
parsedData = [] 

### INPUT FILE along with the path where it it located
with open(r"D:\\DataScience\\In\\Chat Analysis\\WhatsApp Chat - MITRA - 1 ASD Awareness\\WhatsApp Chat - MITRA - 1 ASD Awareness.txt","r",encoding="utf8") as f:

    f=f.readlines()
    messageBuffer = [] 
    date, time, author = None, None, None 
    for line in f:
        line=line.lstrip()
        count3=count3+1
        if len(line)>0:
            count6=count6+1
            if line[1]=='[':
                count5=count5+1
                line=line[1:]
            if startsWithDate(line):
                if count1==0:
                    line11=line
                line22=line
                print('T',line)
                count1=count1+1
                if len(messageBuffer) > 0: 
                    parsedData.append([date, time, author, ' '.join(messageBuffer)]) 
                messageBuffer.clear() 
                date, time, author, message = getDTAM(line) 
                if author == None:
                    count7=count7+1
                messageBuffer.append(message) 
            else:
                print('F',line)
                count2=count2+1
                messageBuffer.append(line) 
        else:
            print('B',line)
            count4=count4+1
```
